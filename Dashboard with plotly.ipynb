```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/content/netflix_titles.csv")
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.columns
```

```python
fig = px.line(df, x='release_year', y='title',color="rating", title="Line graph")
fig.show()
fig.write_html("pc.html1")
```

```python
fig1 = px.line(df, x='release_year', y='duration',color="rating", title="Line graph")
fig1.show()
fig1.write_html("pc.html2")
```

```python
fig2 = px.scatter(df, x='release_year', y='title',color="rating",size=[1.0]*5332, title="Scatter graph")
fig2.show()
fig2.write_html("pc.html3")
```

```python
fig3 = px.box(df, x='release_year', y='duration',color="rating", title="Box graph")
fig3.show()
fig3.write_html("pc.html4")
```

```python
fig4 = px.histogram(df, x='release_year',  title='Histogram of Release Year')
fig4.show()
fig4.write_html("pc.html5")
```

```python
fig5 = px.density_heatmap(df, x='release_year', y='duration', title="Heatmap")
fig5.show()
fig5.write_html("pc.html6")
```
