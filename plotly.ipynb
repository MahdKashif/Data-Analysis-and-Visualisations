```python
import plotly.express as px
# This dataframe has 244 lines, but 4 distinct values for `day`
df = px.data.tips()
fig = px.scatter(df, x='tip', y='day',color="tip")
fig.show()
fig.write_html("pc1.html")
```

```python
import plotly.express as px
# This dataframe has 244 lines, but 4 distinct values for `day`
df = px.data.tips()
fig = px.pie(df, values='tip', names='day',color="size")
fig.show()
fig.write_html("pc2.html")
```

```python
fig = px.line(df, x='Bldg #', y='Side',color="Deaths", title="Line graph")
fig.show()
fig.write_html("pc3.html")
```

```python
fig = px.box(df, x='day', y='tip',color="size",title='Box graph respective of data and tip')
fig.show()
fig.write_html("pc4.html")
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("bird-window-collision-death.csv")
fig = px.pie(df, values='Deaths', names='Bldg #', title="Death with respect to Buildings", color="Side", hole=0.5)
fig.show()
fig.update_traces(insidetextfont_color="white", textinfo="label+percent")
fig.update_layout(legend_itemclick=False)
fig.write_html("pc5.html")
```
